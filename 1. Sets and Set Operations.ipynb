### Intersection

```python
n1 = {n for n in range(1,20,1)}
n2 = {n for n in range(0,20, 2)}
```

```python
n1
```

```python
n2
```

```python
n1.intersection(n2)
```

```python
n1 & n2
```

### Difference

The set of elements that are in A, but not in B

```python
n1.difference(n2)
```

```python
n2.difference(n1)
```

```python
n1 - n2
```

```python
n2 - n1
```

### Symmetric difference

Elements that are in either of the sets but not in their intersection.  Also known as the disjunctive union and set sum.

```python
n1.symmetric_difference(n2)
```

```python
n1 ^ n2
```

### Disjoint

For set A and B.  If the intersection of A and B is the empty set, they are disjoint.  i.e. they ahve nothing in common.

```python
n3 = {555,777,888}
```

```python
n1.isdisjoint(n3) # no elements in common 
```
